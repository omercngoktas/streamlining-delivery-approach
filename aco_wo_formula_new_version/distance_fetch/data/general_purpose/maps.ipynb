```python
import pandas as pd
import folium
```

```python
stores = pd.read_csv("../store/stores.csv")
```

```python
istanbul_latitude = 40.98949
istanbul_longitude = 29.2432
```

```python
# display the map of Istanbul
istanbul_map = folium.Map(location=[istanbul_latitude, istanbul_longitude], zoom_start=12)
```

```python
istanbul_map
```

```python
incidents = folium.map.FeatureGroup()
```

```python
depots = pd.read_csv("../depot/depots.csv")

#loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(stores.latitude, stores.longitude):
    incidents.add_child(
        folium.features.CircleMarker(
            [float(lat), float(lng)],
            radius=5, # define how big you want the circle markers to be
            color='blue',
            fill=True,
            fill_color='red',
            fill_opacity=0.6
        )
    )# add incidents to map

#loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(depots.latitude, depots.longitude):
    incidents.add_child(
        folium.Marker(
            [float(lat), float(lng)],
        )
    )# add incidents to map

istanbul_map.add_child(incidents)
```

```python
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(stores.latitude, stores.longitude):
    incidents.add_child(
        folium.features.CircleMarker(
        [lat, lng],
        radius=5, # define how big you want the circle markers to be
        color='yellow',
        fill=True,
        fill_color='red',
        fill_opacity=0.6
        )
    )# add pop-up text to each marker on the map
 
latitudes = list(stores.latitude)
longitudes = list(stores.longitude)
labels = list(stores.store_location)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(istanbul_map) 
 
# add incidents to map
istanbul_map.add_child(incidents)
```

```python
from folium import plugins# lets start again with a clean copy of the map of San Francisco
istanbul_map = folium.Map(location = [istanbul_latitude, istanbul_longitude], zoom_start = 12)# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(istanbul_map)# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(stores.latitude, stores.longitude, stores.store_location):
 folium.Marker(
 location=[lat, lng],
 icon=None,
 popup=label,
 ).add_to(incidents)# display map
istanbul_map
```
